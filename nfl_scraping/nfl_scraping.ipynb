```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from shutil import which
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_path = which("chromedriver")
driver = webdriver.Chrome(executable_path=chrome_path, options=chrome_options)
driver.set_window_size(1920, 1080)
driver.get("https://www.footballdb.com/scores/index.html")
```

```python
dict = []
for i in range(0,4):
    date_message = driver.find_elements_by_xpath('//div/div/table/thead/tr[2]/th[1]')[i]
    date = date_message.text
    #team scores
    score_message1 = driver.find_elements_by_xpath('//div/div/table/tbody/tr[1]/td[6]/b')[i]
    team_score_1 = score_message1.text
    score_message2 = driver.find_elements_by_xpath('//div/div/table/tbody/tr[2]/td[6]/b')[i]
    team_score_2 = score_message2.text
    #team names
    user_message1 = driver.find_elements_by_xpath('//div/div/table/tbody/tr[1]/td[1]/a')[i]
    team_name_1 = user_message1.text
    user_message2 = driver.find_elements_by_xpath('//div/div/table/tbody/tr[2]/td[1]/a')[i]
    team_name_2 = user_message2.text
    #pass
    pass_message1 = driver.find_elements_by_xpath('//div/div/div/div[1]/div[1]/div[2]/a[1]')[i]
    pass_ply_1 = pass_message1.text
    pass_message2 = driver.find_elements_by_xpath('//div/div/div/div[1]/div[1]/div[2]/a[2]')[i]
    pass_ply_2 = pass_message2.text
    # rush
    rush_message1 = driver.find_elements_by_xpath('//div/div/div/div[1]/div[2]/div[2]/a[1]')[i]
    rush_ply_1 = rush_message1.text
    rush_message2 = driver.find_elements_by_xpath('//div/div/div/div[1]/div[2]/div[2]/a[2]')[i]
    rush_ply_2 = rush_message2.text
    # rec
    rec_message1 = driver.find_elements_by_xpath('//div/div/div/div[1]/div[3]/div[2]/a[1]')[i]
    rec_ply_1 = rec_message1.text
    rec_message2 = driver.find_elements_by_xpath('//div/div/div/div[1]/div[3]/div[2]/a[2]')[i]
    rec_ply_2 = rec_message2.text
    if team_score_1 > team_score_2:
        dic = {
        "game_date":date,
        "winning_team":team_name_1,
        "losing_team":team_name_2,
        "winning_score":team_score_1,
        "losing_score":team_score_2 ,
        "winning_PASS":pass_ply_1,
        "winning_RUSH":rush_ply_1,
        "winning_REC":rec_ply_1,
        "losing_PASS":pass_ply_2,
        "losing_RUSH":rush_ply_2,
        "losing_REC":rec_ply_2
    }
    else:  
        dic = {
        "game_date":date,
        "winning_team":team_name_2,
        "losing_team":team_name_1,
        "winning_score":team_score_2,
        "losing_score":team_score_1,
        "winning_PASS":pass_ply_2,
        "winning_RUSH":rush_ply_2,
        "winning_REC":rec_ply_2,
        "losing_PASS":pass_ply_1,
        "losing_RUSH":rush_ply_1,
        "losing_REC":rec_ply_1 
        }
    dict.append(dic)
dict    
```

```python
import json
with open('test.txt', 'w') as f:
    f.write(json.dumps(dict))
```
