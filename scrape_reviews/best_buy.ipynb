```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
```

```python
HEADERS = {
    'authority': 'scrapeme.live',
    'dnt': '1',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'sec-fetch-site': 'none',
    'sec-fetch-mode': 'navigate',
    'sec-fetch-user': '?1',
    'sec-fetch-dest': 'document',
    'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
}
pages = []
pages_to_scrape= 15
for i in range(1,pages_to_scrape+1):
    url = ('https://www.bestbuy.com/site/reviews/blink-mini-indoor-1080p-wi-fi-security-camera-2-pack-white/6401019?variant=A&sort=MOST_RECENT&page={}').format(i)
    pages.append(url)
```

```python
titles = []
for item in pages:
    page = requests.get(item, headers = HEADERS)
    soup = bs4(page.content, 'lxml')
    for i in soup.findAll('h4',class_='review-title'):
        title=i.getText()
        titles.append(title)
len(titles)        
```

```python
dates = []
for item in pages:
    page = requests.get(item, headers = HEADERS)
    soup = bs4(page.content, 'lxml')
    for d in soup.findAll('div',class_="review-context"):
        tag = soup.time 
        date = tag['title']   
        dates.append(date)
len(dates)        
```

```python
usernames = []
for item in pages:
    page = requests.get(item, headers = HEADERS)
    soup = bs4(page.content, 'lxml')
    for i in soup.findAll('div',class_='ugc-author'):
        names=i.getText()
        usernames.append(names)   
usernames = usernames[1:len(usernames):3]
len(usernames)    
```

```python
reviews = []
for item in pages:
    page = requests.get(item, headers = HEADERS)
    soup = bs4(page.content, 'lxml')
    for r in soup.findAll("div",class_="ugc-review-body"):
        review=r.getText()
        review = review.replace("\n","")
        reviews.append(review)
len(reviews)    
```

```python
stars = []
for item in pages:
    page = requests.get(item, headers = HEADERS)
    soup = bs4(page.content, 'lxml')
    for s in soup.findAll('p',class_='sr-only'):
        rate=s.getText()
        rate = rate[7:23]
        stars.append(rate)
    stars=stars[2:]
len(stars)    
```

```python
#brand name and model_number
product_related = []
for i in soup.findAll("dd"):
    SKU = i.getText()
    product_related.append(SKU)
sku= product_related[2]
model_number = product_related[1]

# product rating
product_rating = soup.findAll("span",class_="ugc-c-review-average")[0]
rating = product_rating.getText()

#product review
product_review = soup.findAll("span",class_="c-reviews-v4")[0]
product_reviews = product_review.getText()
product_reviews = product_reviews[1:4]

#product title
product_titles = soup.findAll("h2",class_="product-title")[0]
product_title = product_titles.getText()

#product price
product_price = soup.findAll("div",class_="priceView-customer-price")[0]
price = product_price.getText()
price = price[-7:]
```

```python
product_rev = []
product_rat = []
SKu = []
MODEL = []
global_reviews = []
product_tit = []
product_price = []
for i in range(len(usernames)):
    SKu.append(sku)
    MODEL.append(model_number)
    product_rat.append(rating)
    product_rev.append(product_reviews)
    product_tit.append(product_title)
    product_price.append(price)
```

```python
df = pd.DataFrame({"Product_Title":product_tit,"Model":MODEL,"SKU":SKu,"Product_rating":product_rat,"Global_reviews":product_rev,"Product_price":product_price,"Date":dates, "Usernames":usernames,"Titles":titles, "Stars":stars, "Reviews":reviews                  
})    
df.index+=1
df
```

```python
df.to_excel("23 bestbuy.xlsx")
```
