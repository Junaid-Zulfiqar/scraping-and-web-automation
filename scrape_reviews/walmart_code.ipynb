```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
import numpy as np
```

```python
usernames=[]
stars=[]
titles=[]
customer_reviews=[]
dates = []
#review page url
url = ('https://www.walmart.com/reviews/product/479742211?page=3')
page = requests.get(url)
soup = bs4(page.content, 'html.parser')
#username
username = soup.find_all("span",{'itemprop':'author'})
for i in range(0,len(username)):
    usernames.append(username[i].get_text())
#title
title = soup.find_all('h3',class_='review-title')
for i in range(0,len(title)):
    titles.append(title[i].get_text())
#date
date = soup.find_all('span',{'itemprop':'datePublished'})
for d in range(0,len(date)):
    dates.append(date[d].get_text())
#stars
stars=[]
for s in soup.findAll('span',class_='seo-avg-rating'):
    rate=s.getText()
    stars.append(rate)
stars = stars[2:]  
stars.append(5)
#review
review = soup.find_all('div',class_='review-text')
for t in range(0,len(review)):
    customer_reviews.append(review[t].get_text())
```

```python
#Dataframe
df = pd.DataFrame({"Date":dates,"Name":usernames,"Stars":stars,"Titles":titles, "Reviews":customer_reviews})
df.index+=1
df
```

```python
product_title = soup.findAll("h1")[0]
product_title = product_title.getText()
product_title
```

```python
product_review = soup.findAll("span",class_="stars-reviews-count-node")[0]
product_reviews = product_review.getText()
product_reviews = product_reviews.split(" ")
product_reviews = product_reviews[0]
product_reviews
```

```python
product_rating = soup.find("span",{"role":"link"})['aria-label']
product_rating = product_rating[0:4]
product_rating
```

```python
product_price = soup.find("span",{"itemprop":"price"})
product_price = product_price.getText()
product_price
```

```python
product_brand = soup.find("div",class_="copy-mini")
product_brand = product_brand.getText()
product_brand
```

```python
product_model = soup.find("div",class_="other-info")
product_model = product_model.getText()
product_model = product_model.split(":")
product_model = product_model[1]
product_model
```

```python
walmart_number = soup.find("div",class_="wm-item-number")
walmart_number = walmart_number.getText()
walmart_number = walmart_number.split("#")
walmart_number = walmart_number[1]
walmart_number
```
