```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
import numpy as np
```

```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
URL = "https://www.ebay.com/p/13021690289?iid=303749094088"
webpage = requests.get(URL, headers=HEADERS)
soup = bs4(webpage.content, "lxml")
```

```python
print(soup.prettify())
```

```python
usernames = []
for i in soup.findAll("a",class_="review-item-author"):
    names=i.getText()
    usernames.append(names)   
usernames   
```

```python
titles = []
for i in soup.findAll("h3",{"itemprop":"name"}):
    title=i.getText()
    titles.append(title)   
titles   
```

```python
dates=[]
for d in soup.findAll('span',class_='review-item-date'):
    date=d.getText()
    dates.append(date)
dates  
```

```python
stars = []
for s in soup.findAll('span', class_='star-rating'):
        for k,v in s.attrs.items():
            star =v[:1]
            stars.append(star)
stars = stars[2:len(stars):3]    
stars =stars[1:]
stars
```

```python
# reviews        
reviews = []
for r in soup.findAll("p",{"itemprop":"reviewBody"}):
    review=r.getText()
    review = review.replace("\n","")
    review = review.replace("Read more","")
    reviews.append(review)   
reviews    
```

```python
df = pd.DataFrame({"Date":dates, "Usernames":usernames, "Stars":stars,"Titles":titles, "Reviews":reviews})    
df.index+=1
df
```

```python
list = [len(usernames),len(dates),len(stars),len(reviews),len(titles)]
print(list)
```

```python
df.to_csv("ebay80.csv")
```
