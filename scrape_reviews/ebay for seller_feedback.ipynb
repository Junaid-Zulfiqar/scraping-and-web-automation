```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
```

```python
url = 'https://www.ebay.com/fdbk/feedback_profile/u-smarthome?filter=feedback_page:RECEIVED_AS_SELLER&_trksid=p2047675.l2560'
page = requests.get(url)
soup = bs4(page.content, "lxml")
```

```python
#title
titles = []
for i in soup.findAll('div',class_="card__comment"):
    title=i.getText()
    titles.append(title) 
# price
prices=[]
for s in soup.findAll('div',class_="card__price"):
    rate=s.getText()
    prices.append(rate)
#buyer
usernames=[]
for s in soup.findAll('span',{"aria-label":"Feedback left by buyer."}):
    name=s.getText()
    usernames.append(name)
#buyer
stars=[]
for s in soup.findAll('span',class_="no-wrap"):
    rate=s.getText()
    stars.append(rate)    
#reviews
reviews = []
for r in soup.findAll("div",class_="card__item"):
    review=r.getText()
    reviews.append(review)
#date
dates=[]
datted = 'past 6 months'
for d in soup.findAll('span',{"aria-label":"Past month"}):
    date=d.getText()
    dates.append(date)
while(len(dates)<len(usernames)):
    dates.append(datted)
```

```python
df = pd.DataFrame({"usernames": usernames, "rating":stars,"title" :titles,"Reviews":reviews, "Date":dates                 
})    
df.index+=1
df
```
